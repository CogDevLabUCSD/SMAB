```python
# Preprocessing pipeline:
# 1. separate game 1 and game 2
#(row num - 2)
# if single game: (4 x 100)
# 2. remove confederate trials (even trials)
# 3. rows needed:
#     trial number (row 2)
#     probability of the arm chosen (row 5)
#     reward or no (row 6)
#     current score (row 8)
    
# if joint game: (5 x 200)
# 2. keep the partner's trials
# 3. rows needed:
#     trial number (row 2)
#     prob of the arm chosen (row 5)
#     reward or no (row 6)
#     current score player 1 (row 8)
#     current score player 2 (row 9)
```

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/Users/yubinghong/Desktop/data/s01_gamedata.csv') #change file name 
print(data.shape)
data.head()
```

```python
# separate game 1 and game 2
game1_df = data.iloc[:, 0:200]
print(game1_df.head())
print(game1_df.shape)

game2_df = data.iloc[:, 200:400]
print(game2_df.head())
print(game2_df.shape)
```

# for single player game


```python
#game1
#remove confederate trials (even trials)
single_df = game1_df.iloc[:, [i for i in range(len(game1_df.columns)) if i%2 == 0]]
print(single_df.shape)

#choose row 
single_df = single_df.iloc[[0,3,4,6]]
print(single_df.shape)
single_df
```

```python
#game 2

#remove confederate trials (even trials)
single_df = game2_df.iloc[:, [i for i in range(len(game2_df.columns)) if i%2 == 0]]
print(single_df.shape)

#choose row
single_df = single_df.iloc[[0,3,4,6]]
print(single_df.shape)
single_df
```
